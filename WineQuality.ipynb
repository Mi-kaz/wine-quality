```python
from sklearn.model_selection import train_test_split

rdata = red_df[['alcohol', 'volatile acidity', 'sulphates']].values
rlabels = red_df['quality'].values

rdata_train, rdata_test, rlabels_train, rlabels_test = train_test_split(rdata,
                                                                    rlabels,
                                                                    test_size = 0.33,
                                                                    random_state = 10)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state = 10)

rdata_train, rlabels_train = smote.fit_resample(rdata_train, rlabels_train)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(rdata_train)
rdata_train_scaled = scaler.transform(rdata_train)
rdata_test_scaled = scaler.transform(rdata_test)
```

```python
from sklearn import tree
from sklearn.metrics import accuracy_score

rdecision_tree = tree.DecisionTreeClassifier(random_state = 10)
rdecision_tree.fit(rdata_train_scaled, rlabels_train)

decision_tree_predictions = rdecision_tree.predict(rdata_train_scaled)
train_acc = accuracy_score(rlabels_train, decision_tree_predictions)

decision_tree_predictions = rdecision_tree.predict(rdata_test_scaled)
test_acc = accuracy_score(rlabels_test, decision_tree_predictions)
```

```python
print(f"Decision tree train acc: {train_acc * 100}%")
print(f"Decision tree test acc: {test_acc * 100}%")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error


rlinear_regression = LinearRegression()
rlinear_regression.fit(rdata_train_scaled, rlabels_train)

rlinear_regression_predictions = rlinear_regression.predict(rdata_train_scaled)
train_mae = mean_absolute_error(rlabels_train, rlinear_regression_predictions)

rlinear_regression_predictions = rlinear_regression.predict(rdata_test_scaled)
test_mae = mean_absolute_error(rlabels_test, rlinear_regression_predictions)
```

```python
print(f"Linear regression train mae: {train_mae}")
print(f"Linear regression test mae: {test_mae}" )
```
